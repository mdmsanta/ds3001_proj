```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("clean.csv")
```

```python
from collections import Counter

#iterate through the words and collect the top 30 most common words
for item in [data.commonFileTypes]: 
    c = Counter(item)
    piedata = c.most_common()[:30]
```

```python
print(piedata)
```

```python
plt.pie([8038,4470,470,42], labels= ["csv","other","json","sqlite"],
        shadow=True,autopct='%1.1f%%', radius = 1)
plt.title("Data Type Frequency")
plt.show()
```

```python
for item in [data.currentDatasetVersionNumber]:
    c = Counter(item)
    plt.hist(list(c.values()),bins=[0, 5, 10, 15, 20, 25, 30, 35, 40, 50, 100])
    
#style the graph    
plt.title("Number of data sets versions")
plt.xlabel("Current Data Set Version")
plt.ylabel("Number of Datasets")
plt.show()
```

```python
true  = 0
false  = 0 
for thing in data.isFeatured:
    if thing == True:
        true = true +1
    elif thing == False:
        false = false +1 
```

```python
print(false)
```

```python
print(true)
```

```python
plt.pie([true, false], labels = ["Is Featured", "Isn't Featured"], shadow = True, autopct='%1.1f%%')
plt.title("Data Sets Featured or Not")
```

```python
true2  = 0
false2  = 0 
for thing in data.isSuperFeatured:
    if thing == True:
        true2 = true2 +1
    elif thing == False:
        false2 = false2 +1 
```

```python
plt.pie([true2, false2], labels = ["Is Super Featured", "Isn't Super Featured"], shadow = True, autopct='%1.1f%%')
plt.title("Data Sets Super Featured or Not")
```

```python
plt.scatter(data.viewCount, data.downloadCount, s = 1.5)
plt.title("View versus Download Count")
plt.xlabel("Veiw Count")
plt.ylabel("Download Count")
```

```python
ratio = []
download = data.downloadCount
view = data.viewCount
for i in range(0, len(download)):
    if view[i]==0:
        val=0
    else:
        val = download[i]/view[i]
    ratio.append(val)
```

```python
plt.scatter(ratio, data.viewCount, s = 1.5)
plt.title("Ratio versus Vote Button")
plt.xlabel("Ratio")
plt.ylabel("Vote Button")
```
